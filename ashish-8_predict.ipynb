```python
import numpy as np
from keras.preprocessing import image
```

```python
from keras.models import load_model
```

```python
model=load_model('classify-8_2.h5')
```

```python
model.summary()
```

thus we build and imported  it its time on predict some new images


the outputs for <ul><li>aeroplane will be--0</li><li>car will be--1</li><li>cat will be--2</li><li>dog will be--3</li><li>flower will be --4</li><li>fruit will be--5</li><li>motorbike will be--6</li><li>person will be--7</li></ul>

```python
import matplotlib.pyplot as plt
```

```python
img=image.load_img('predict/man3.jpg',target_size=(100,100))

plt.imshow(img)
plt.axis('off')
plt.show()
img_tensor=image.img_to_array(img)
img_tensor=np.expand_dims(img_tensor,axis=0)
img_tensor/=255
```

```python
model.predict_classes(img_tensor)
```
