<h1> Keras build in VGG-16  </h1>

In 2014 deep convolutional network was developed by Visual Geometry group (VGG) for object detection .they made 16-19 convolutional layers and trained on Imagenet images which is having 1000 classes .the model was trained on 1.3 million images ,50000 validation and 100000 testing images .

later on weights learned by this model on caffe have been converted into keras



```python
from keras.applications.vgg16 import VGG16
```

```python
model=VGG16(weights='imagenet',include_top=True)
```

```python
model.summary()
```

```python
import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
from keras.applications.vgg16 import preprocess_input,decode_predictions
```

```python
img=image.load_img('predict/dog.jpg',target_size=(224,224))

plt.imshow(img)
plt.axis('off')
plt.show()
x=image.img_to_array(img)
x=np.expand_dims(x,axis=0)
x=preprocess_input(x)
pred=model.predict(x)
print("i think it's ",decode_predictions(pred,top=3)[0])
```
